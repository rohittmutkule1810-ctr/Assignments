# Hypothesis Testing


 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?
    ->
    The null hypothesis is the claim that there is no effect in the population. Researcher try to disprove the hypothesis.

    The null hypothesis is importance because its default assumption that experiments has no effect, which provides a starting point for testing.


2: What does the significance level (α) represent in hypothesis testing?
    ->
    
    In hypothesis testing, the significance level (α) represents the probability of making a Type I error.


3: Differentiate between Type I and Type II errors.
    ->
    Type-1 error - Probability of rejecting the null hypothesis when it is true.
    Denonated as a (α).
    Also known as alpha error.
    Also known as false positive.
    Refered as error first kind.
    Error can be reduced by reducing the alpha level.

    Type-2 error - Probability of accepting the null hypothesis when it is false.
    Denonated as a (β).
    Also known as beta error.
    Also known as false negative.
    Refered as error second kind.
    Error can be reduced by increasing the alpha level.


4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.
    ->
    One tailed- One tailed is unidirectional.
    critical region for rejecting Ho is located in only one tail of the ditribution.
    Research hypothesis is directional greater than or less than.
    Ex.
    Right-tailed test - μ > 100
    Testing if the mean is greater than 100.
    Left-tailed test - μ < 100
    Testing if the mean is less than 100.

    Two Tailed- Two directional is bi-directional.
    critical region for rejecting Ho is divided between both tails of the ditribution.
    Research hypothesis is non-directional not equal to.
    Ex.
    μ != 100
    Testing if the mean is different ( higher OR lower) than 100.
