2: What does a positive, negative, and zero covariance indicate about the relationship between two variables?
    
    
    ->
    Positive Covariance - When covariance is positive, it means both variables move in the same direction.If one increases, the other also tends to increase.
    Same Direction Relationship.
    Negative Covariance - A negative covariance means the variables move in opposite directions.If one increases, the other tends to decrease.
    Opposite Direction Relationship.
    Zero Covariance - Zero covariance means the variables do not show any linear relationship.Their movements are unrelated
    No Linear Relationship.


3: Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?
    
    
    ->
    Limitations of Covariance -
    Depends on Units of Measurement
    Covariance is expressed in the combined units of the two variables.
    No Fixed Range
    Covariance can take any value, so we cannot tell whether a relationship is weak or strong just by looking at the number.
    Difficult to Compare
    the covariance from one dataset cannot be directly compared with another dataset.
    Shows Only Direction
    Covariance only tells whether variables move in the same or opposite direction.

    Correlation Preferred
    -Correlation is Sta.ndardized
    -Easy to Interpret.
    -Shows Both Strength and Direction.
    -Comparable Across Variables & Datasets.

5: If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?
    
    
    ->
    Interpretation of r = 0.85
    A correlation coefficient of 0.85 indicates a strong positive linear relationship between variables X and Y.
    This means -
    When X increases, Y also tends to increase.
    The points lie close to a straight upward-sloping line.
    The relationship is strong and consistent.
    Infer Causation-
    No.
    A correlation value of 0.85 does NOT prove that X causes Y.
    Why not-
    Correlation only shows association, not cause-and-effect.
    There might be a third variable affecting both X and Y.
    The relationship could be coincidental.
